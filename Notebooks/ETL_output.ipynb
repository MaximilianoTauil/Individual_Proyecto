### Transformación de datos

Se creó un DataFrame con todas las filas nulas en su totalidad para ver cómo se las iba a tratar.

Se observó que hay columnas las cuales no son necesarias para nuestra API, serán eliminadas. Ellas son: 
* Url
* Reviews_url
* Tags
* Specs

```python
data_output = data_output.drop(['url','reviews_url','tags','specs'],axis=1)
data_output
```

La columna "release_date" tiene valores NaN y formato de fecha, solo es necesario el año. Para extraerlo, se utilizó una función de nuestro módulo Herramientas. 

```python
data_output['release_date'].fillna('No date',inplace=True)
data_output['release_date'] = data_output['release_date'].apply(Herr.extraccion_anio_wstring)
data_output
```

Se revisó la columna "id" para ver si había duplicados, y efectivamente hay tres usuarios, dos de ellos con id NaN los cuales serán eliminados y uno que está duplicado, el cual también será eliminado.

```python
duplicadas_id = data_output[data_output.duplicated(subset='id',keep=False)]
duplicadas_id_orden = duplicadas_id.sort_values(by='id')
duplicadas_id_orden
```

Se eliminaron los duplicados a través de su índice.

```python
dropear = [102204,88384,119271]
data_output = data_output.drop(dropear)
```

Se verificó que columnas faltan a tratar.

```python
Herr.analizar_datos(data_output)
```

Mediante la visualización de los tipos de datos, vemos que "price" tiene tipos float y string. Lo que corresponde es que sean números o, si es gratis, marcar "price = 0".

```python
data_output['price'].unique()
```

Se cambió todo lo que no eran números en la columna "price", ya sea porque son gratis u otras palabras que se encontraron, se pasaron a números para poder utilizarlos.

```python
strings_price = ['Free To Play','Free to Play','Free','Free Demo','Play for Free!','Install Now','Play WARMACHINE: Tactics Demo','Free Mod','Install Theme','Third-party','Play Now','Free HITMAN™ Holiday Pack','Play the Demo','Free to Try','Free Movie','Free to Use']
data_output['price'] = data_output['price'].replace(strings_price,0)
data_output['price'] = data_output['price'].replace('nan',None)
data_output['price'] = data_output['price'].fillna(0)
data_output = data_output[data_output['price'].apply(lambda x: not str(x).startswith('Starting at'))]
data_output['price'] = data_output['price'].astype(float)
```

Se reseteó el índice del DataFrame porque se realizaron muchos cambios.

```python
data_output = data_output.reset_index(drop=True)
```

```python
data_output = data_output.explode('genres')
data_output
```

```python
data_output = data_output.dropna(subset='genres')
```

Se pasó una lista de columnas con valores NaN para rellenarlos con "No data", dado que no podemos rellenarlos con ningún valor. Esto sucede porque son nombres de los juegos, desarrolladores, géneros, entre otros.

```python
columnas_Nan = ['publisher','genres','app_name','title','developer','release_date']
data_output[columnas_Nan] = data_output[columnas_Nan].fillna('No data')
```

Hacemos una última revisión para corroborar los tipos de datos y los valores nulos.

```python
Herr.analizar_datos(data_output)
```
